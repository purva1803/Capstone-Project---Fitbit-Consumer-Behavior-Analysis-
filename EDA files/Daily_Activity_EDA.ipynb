# **Data Visualization**


**Bar Plot :**

 In this I analyze activity data to identify the top 10 users based on the maximum distance covered in a single day. The provided dataset contains information about users' activities, including their ID, activity date, and total distance covered. By aggregating this data and visualizing it using a bar plot,I was able to identify the users who have achieved the highest maximum distance covered in a single day.Based on data it seems user 1624580081 has coverd maximum steps.

```python
# Aggregating data by user ID and activity date to compute total distance traveled
daily_distance = df.groupby(['Id', 'ActivityDate'])['TotalDistance'].sum().reset_index()

# Determining the peak distance achieved by each user within a day
max_dist_per_day = daily_distance.groupby('Id')['TotalDistance'].max().reset_index()
top_10_users = max_dist_per_day.sort_values(by='TotalDistance', ascending=False).head(10)

plt.figure(figsize=(12, 8))
sns.barplot(data=top_10_users, x='Id', y='TotalDistance', color='blue', order=top_10_users['Id'])
plt.title('Top 10 Users by Maximum Distance Covered in a Single Day')
plt.xlabel('User ID')
plt.ylabel('Maximum Distance Covered in a Single Day')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Aggregating data by user ID and activity date to compute total calories traveled
daily_distance = df.groupby(['Id', 'ActivityDate'])['Calories'].sum().reset_index()
max_dist_per_day = daily_distance.groupby('Id')['Calories'].max().reset_index()
top_10_users = max_dist_per_day.sort_values(by='Calories', ascending=False).head(10)

plt.figure(figsize=(12, 8))
sns.barplot(data=top_10_users, x='Id', y='Calories', color='blue', order=top_10_users['Id'])
plt.title('Top 10 Users by Maximum Calories Covered in a Single Day')
plt.xlabel('User ID')
plt.ylabel('Maximum Calories Covered in a Single Day')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Pie-Chart:**

It categorizes users into different activity levels based on their activity metrics such as distances covered and active minutes spent. The summary is presented as a pie chart, with each slice representing a specific activity level. Activity levels include Highly Active, Moderately Active, Mildly Active, Sedentary Active, and Inactive. The chart offers insights into the distribution of users across these activity levels, aiding in understanding overall activity patterns within the dataset.



```python
# Define a function to classify activity level based on distances covered and active minutes spent
def classify_activitylevel(row):
    if row['VeryActiveDistance'] > 0 or row['VeryActiveMinutes'] > 35:
        return 'Highly Active'
    elif row['ModeratelyActiveDistance'] > 0 or row['FairlyActiveMinutes'] > 35:
        return 'Moderately Active'
    elif row['LightActiveDistance'] > 0 or row['LightlyActiveMinutes'] > 35:
        return 'Mildly Active'
    elif row['SedentaryActiveDistance'] > 0 or row['SedentaryMinutes'] > 35:
        return 'Sedentary Active'
    else:
        return 'Inactive'

# Applying the function to create a new 'activitylevel' column
df['activitylevel'] = df.apply(classify_activitylevel, axis=1)

# Creating a new DataFrame to store the summarized results
activitylevel_summary = pd.DataFrame(columns=['Id', 'activitylevel', 'TotalSteps'])

# Iterating over unique IDs to find the top user for each activitylevel
for user_id in df['Id'].unique():
    # Extracting data for the current user
    user_data = df[df['Id'] == user_id]

    # Finding the top user for each activitylevel for the current user
    top_user_per_activitylevel = user_data.groupby('activitylevel').apply(lambda x: x.nlargest(1, 'TotalSteps'))[['Id', 'activitylevel', 'TotalSteps']]

    # Appending the top result to the new DataFrame
    activitylevel_summary = activitylevel_summary.append(top_user_per_activitylevel, ignore_index=True)

# Displaying the top user for each activitylevel overall
print(activitylevel_summary)

import matplotlib.pyplot as plt

# Calculating the counts of each activitylevel
activitylevel_counts = activitylevel_summary['activitylevel'].value_counts()


# Plotting the pie chart with custom colors
plt.figure(figsize=(8, 8))
plt.pie(activitylevel_counts, labels=activitylevel_counts.index, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('Activity Level Summary')
plt.axis('equal')  # Ensuring the pie is drawn as a circle
plt.show()
```
